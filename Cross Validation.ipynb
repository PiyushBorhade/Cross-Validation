```python
import pandas as pd
```

```python
df = pd.read_csv('diabetes.csv')
```

```python
df
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
X = df.drop('Outcome',axis=1)
y = df['Outcome']
```

### Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_sc = sc.fit_transform(X)
```

### Cross Validation 

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X_sc,y,random_state=11,test_size=0.2)
```

```python
print(X_train.shape)
print(X_test.shape)
```

### KNN  

```python
from sklearn.neighbors import KNeighborsClassifier
clf = KNeighborsClassifier()
```

```python
clf.fit(X_train,y_train)
```

```python
clf.score(X_test,y_test)
```

### K-fold technique 

```python
from sklearn.model_selection import KFold
Kf = KFold(n_splits=5)
```

```python
# for checking whether i and j are given to training and testing resp.

for i,j in Kf.split(X_sc,y):
    print(i,j)
    break
```

```python
154/614
```

25% of data is used in testing and other 75% of data is used in training

```python
from sklearn.neighbors import KNeighborsClassifier
clf = KNeighborsClassifier()
```

```python
list1 = []

for i,j in Kf.split(X_sc,y):
    X_train,X_test,y_train,y_test = X_sc[i],X_sc[j],y[i],y[j]
    clf.fit(X_train,y_train)
    score = clf.score(X_test,y_test)
    list1.append(score)
```

```python
list1
```

```python
import numpy as np
```

```python
accuracy = sum(list1)/5
accuracy
```

The Accuracy we get is around 72.9% i.e 73% manually

###  Cross Val Score

```python
from sklearn.model_selection import cross_val_score
```

```python
acc = cross_val_score(clf,X_sc,y,cv=5)
acc
```

```python
np.mean(acc)
```

The Accuracy we get is 73%

### 1) Gridsearch CV 

```python
from sklearn.svm import SVC
clf = SVC()
```

```python
d = {'C':[1,2,3,4,5,6,7],'kernel':['rbf','ploy','linear'],'gamma':['auto','scale']}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
gd = GridSearchCV(clf,d,cv=5)
```

```python
gd.fit(X_sc,y)
```

```python
gd.cv_results_
```

```python
# convert dictionary into DataFrame
result = pd.DataFrame(gd.cv_results_)
result
```
